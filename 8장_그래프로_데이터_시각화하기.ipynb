````python
import matplotlib.pyplot as plt

sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

# plt.rc('font', family= 'Malgun Gothic')   # 윈도우에서 주피터 노트북 사용시 
# plt.rc('font', family= 'AppleGothic')   # 맥에서 주피터 노트북 사용시 
plt.rc('font', family= 'NanumGothic')   # 구글 코랩 사용시 
plt.title('월별 판매 실적')    # 그래프 제목 넣기
plt.plot(month, sales)
plt.show()
````

````python
plt.rc('font', family='NanumGothic') 
plt.title('월별 판매 실적')
plt.plot(month, sales, color='y') # 그래프 색 바꾸기 
plt.show()
````

````python
plt.figure(dpi=150) # 그래프 해상도 설정하기 
# plt.figure(figsize=(8, 4)) 
plt.rc('font', family='NanumGothic') 
plt.title('월별 판매 실적') 
plt.plot(month, sales, color='b') 
plt.show()
````

````python
plt.rc('font', family='NanumGothic') 
plt.title('월별 판매 실적') 

# 범례설정 label
plt.plot(month, sales, color='b', label='단위: 개수') 
plt.legend()
plt.show()
````

````python
plt.rc('font', family='NanumGothic') 
plt.title('월별 판매 실적') 

plt.plot(month, sales, color='b', label='단위: 개수') 
plt.legend(loc = 'lower right')
plt.show()
````

````python
plt.rc('font', family='NanumGothic') 
plt.title('월별 판매 실적') 
plt.plot(month, sales, color='b', label='단위: 개수') 
plt.legend(loc = 'lower right')
plt.xlabel('월')
plt.ylabel('매출')
plt.show()
````

## 8.3 그래프 겹쳐 그리기
#### 8.3.1 꺾은선 그래프 겹쳐 그리기

````python
import matplotlib.pyplot as plt

singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

plt.plot(singer, week1)
plt.plot(singer, week2)
plt.show()
````

````python
plt.rc('font',family='NanumGothic')
plt.title('오디션 프로그램 득표 현황')
plt.plot(singer, week1, label = '첫째주', color='hotpink')
plt.plot(singer, week2, label = '둘째주', color='royalblue')
plt.legend()
plt.show()
````

#### 8.3.2 막대그래프로 겹쳐 그리기

````python
for i in range(len(week1)) : 
    week1[i] = -week1[i]

plt.rc('font', family='NanumGothic')
plt.title('오디션 프로그램 득표 현황')
plt.barh(singer, week1, label = '첫째주', color = 'hotpink')
plt.barh(singer, week2, label = '둘째주', color = 'royalblue')
plt.legend()
plt.show()
````

````python
for i in range(len(week1)) : 
    week1[i] = -week1[i]

plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 표시
plt.rc('font', family='NanumGothic')
plt.title('오디션 프로그램 득표 현황')
plt.barh(singer, week1, label = '첫째주', color = 'hotpink')
plt.barh(singer, week2, label = '둘째주', color = 'royalblue')
plt.legend()
plt.show()
````

````python
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

plt.rc('font',family='NanumGothic')
plt.title('오디션 프로그램 득표 현황')
plt.bar(range(1,6), week1, label = '첫째주', color='hotpink')
plt.bar(range(7,12), week2, label = '둘째주', color='royalblue')
plt.xticks(range(1, 12), ['A', 'B', 'C', 'D', 'E', ' ', 'A', 'B', 'C', 'D', 'E'])
plt.legend()
plt.show()
````

#### 8.3.3 산점도 겹쳐 그리기

````python
import random

height1, weight1, height2, weight2 =[], [], [], []
for i in range(100) : 
    height1.append(random.randint(140,180))
    weight1.append(random.randint(40,80))
    height2.append(random.randint(160,200))
    weight2.append(random.randint(50,100))
    
plt.rc('font', family='NanumGothic')
plt.title('키와 몸무게의 상관관계')
plt.scatter(height1, weight1, color = 'crimson', alpha = 0.7, label='그룹1')
plt.scatter(height2, weight2, color = 'indigo', alpha = 0.7, label='그룹2')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
````

````python
height1, weight1, score1, height2, weight2, score2 = [], [], [], [], [], [] 

for i in range(100):
    height1.append(random.randint(140, 180)) 
    weight1.append(random.randint(40, 80)) 
    score1.append(random.randint(0, 200)) 
    height2.append(random.randint(160, 200))
    weight2.append(random.randint(50, 100)) 
    score2.append(random.randint(0, 200))

plt.rc('font', family='NanumGothic')
plt.title('키와 몸무게, 체육 점수의 상관관계')
plt.scatter(height1, weight1, s=score1, color='crimson', alpha=0.7, label='그룹1') 
plt.scatter(height2, weight2, s=score2, color='indigo', alpha=0.7, label='그룹2') 
plt.xlabel('키')
plt.ylabel('몸무게')
plt.legend()
plt.show()
````
